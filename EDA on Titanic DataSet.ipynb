## Data Cleaning 

```python
sns.boxplot(x='Pclass',y='Age',data=train)
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 37

        elif Pclass == 2:
            return 29

        else:
            return 24

    else:
        return Age
```

```python
train['Age']= train[['Age','Pclass']].apply(impute_age,axis=1)
sns.heatmap(train.isnull(),yticklabels=False,cbar=False)
train.drop('Cabin',axis=1,inplace=True)
```

```python
pd.get_dummies(train['Embarked'],drop_first=True).head()
```

```python
sex = pd.get_dummies(train['Sex'],drop_first=True)
embark = pd.get_dummies(train['Embarked'],drop_first=True)
train.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
train.head()
```

```python
train=pd.concat([train,sex,embark],axis=1)
train.head()
```

# # # Logistics Regression 

```python
train.drop('Survived',axis=1).head()
```

```python
train['Survived'].head()
```

```python
from sklearn.model_selection import train_test_split #calling test train split
```

```python
#this will give 4 output parameters syntax is #train_test_split(*arrays, **options)
X_train, X_test, y_train, y_test = train_test_split(train.drop('Survived',axis=1), 
                                                    train['Survived'], test_size=0.30, 
                                                    random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression() #Making a classifier for the model.
logmodel.fit(X_train,y_train)   #fitting the data in the created classifier
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
accuracy=confusion_matrix(y_test,predictions)
accuracy
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,predictions)
accuracy
```
